## Importing required libraries

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from datetime import datetime
from sklearn.preprocessing import LabelEncoder
import seaborn as sb
```

```python
df = pd.read_csv('100 Sales Records.csv')
```

```python
df
```

```python
usold = df.sort_values(ascending=False,by=['Units Sold'])
usold
```

```python
#d = datetime.strptime(str, '%m-%d-%Y')
#d.strftime("%d/%m/%Y")
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'],format='%Y-%m-%d')
df['Year'] = pd.DatetimeIndex(df['Order Date']).year
df['Year']
```

```python
df['Month'] = pd.DatetimeIndex(df['Order Date']).month
df['Month']
```

```python
df.groupby('Item Type').sum()
```

```python
year_rev =  df.groupby('Year').sum()['Total Revenue'].reset_index()
year_rev
```

```python
plt.figure(figsize=(20,5))
plt.bar(year_rev['Year'],year_rev['Total Revenue'],color='maroon')
plt.xticks(year_rev['Year'])
plt.ylabel('Revenue in Crores')
plt.xlabel('Year')
plt.show()
```

```python
month_rev =  df.groupby('Month').sum()['Total Revenue'].reset_index()
month_rev
```

```python
plt.figure(figsize=(20,5))
plt.plot(month_rev['Month'],month_rev['Total Revenue'],color='maroon')
plt.xticks(month_rev['Month'])
plt.ylabel('Revenue in Crores')
plt.xlabel('Month')
plt.show()
```

```python
unit_rev =  df.groupby('Item Type').sum()['Units Sold'].reset_index()
unit_rev
```

```python
plt.figure(figsize=(30,10))
plt.pie(unit_rev['Units Sold'],labels = unit_rev['Item Type'], autopct ='%0.1f%%', pctdistance = 0.8)
plt.axis("equal")
plt.show()
```

```python
region_rev =  df.groupby('Region').sum()['Units Sold'].reset_index()
region_rev
```

```python
plt.figure(figsize=(30,10))
plt.pie(region_rev['Units Sold'],labels = region_rev['Region'], autopct ='%0.1f%%', pctdistance = 0.8)
plt.axis("equal")
plt.show()
```
